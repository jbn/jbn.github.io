Today I learned that [``git``](https://git-scm.com/) will emit a list of files matching your [``.gitignore`` file](https://git-scm.com/docs/gitignore) if you execute,

```python
!git clean -Xn
```

To **remove** those files, execute,

```python
!git clean -Xf
```

I'm almost always using git, so this saves me from writing a ``.PHONY: clean`` ``Makefile`` rule.

[Learn more about ``git clean`` here.](https://git-scm.com/docs/git-clean)
